```python
import pymongo as mg
import pandas as pd
client = mg.MongoClient(host='mongodb://localhost:27017')
```

```python
database = client['db_NHIS']
```

```python
collection = database['Join Collection NSC2_BND_M20']
```

```python
cursor = collection.find({'SICK_SYM1':'I10'}) # 'SICK_SYM1':'I10' 고혈압 환자만 SELECT 
```

```python
list_BM = list(cursor)
list_BM
```

```python
df_BM = pd.DataFrame(list_BM)
df_BM
```

```python
df_BM_HP=df_BM[['EDC_SBA','BTH_YYYY', 'DTH_YYYYMM', 'MCARE_SUBJ_CD', 'OPRTN_YN', 'TOT_PRSC_DD_CNT', 'MCARE_RSLT_TYPE', 'STD_YYYY' ]]
df_BM_HP
```

```python
df_BM_HP.to_csv('../datasets/Join Collection NSC2_BND_M20.csv')
```
